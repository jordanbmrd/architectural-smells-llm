```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
files = {
    'aioesphomeapi': 'Aioesphomeapi_Dataset.csv',
    'homeassistant-powercalc': 'Homeassistant-powercalc_Dataset.csv',
    'optuna': 'Optuna_Dataset.csv',
    'transformers': 'Transformers_Dataset.csv',
    'celery': 'celery_Dataset.csv',
    'jumperserver': 'jumpserver_Dataset.csv',
    'locust': 'locust_Dataset.csv',
    'openai-python': 'openai-python_Dataset.csv',
    'python-telegram-bot': 'python-telegram-bot_Dataset.csv',
    'qlib': 'qlib_Dataset.csv',
    'supervision': 'supervision_Dataset.csv',
    'yt-dlp': 'yt-dlp_Dataset.csv'
}
palette = sns.color_palette('tab10', 20)
datasets = {name: pd.read_csv(path) for name, path in files.items()}
```

```python
for project_name, df in datasets.items():
    smells = df.columns.drop('Release version')
    plt.figure(figsize=(14, 7))
    for idx, smell in enumerate(smells):
        color = palette[idx % len(palette)]
        rolling_mean = df[smell].rolling(window=3).mean()
        plt.plot(rolling_mean, linestyle='-', label=f"{smell} Moving Average", color=color)
    plt.title(f"{project_name} - Smells with Moving Averages")
    plt.xlabel('Release Index')
    plt.ylabel('Count')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
