```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = {
    'aioesphomeapi': 'Aioesphomeapi_Dataset.csv',
    'homeassistant-powercalc': 'Homeassistant-powercalc_Dataset.csv',
    'optuna': 'Optuna_Dataset.csv',
    'transformers': 'Transformers_Dataset.csv',
    'celery': 'celery_Dataset.csv',
    'jumperserver': 'jumpserver_Dataset.csv',
    'locust': 'locust_Dataset.csv',
    'openai-python': 'openai-python_Dataset.csv',
    'python-telegram-bot': 'python-telegram-bot_Dataset.csv',
    'qlib': 'qlib_Dataset.csv',
    'supervision': 'supervision_Dataset.csv',
    'yt-dlp': 'yt-dlp_Dataset.csv'
}
dataset_df = {name: pd.read_csv(path) for name, path in files.items()}
```

```python
# Chaque projet est un DataFrame dans le dict
dataset_counts = {
    name: df.drop(columns=['Release version'])
    for name, df in dataset_df.items()
}
```

```python
for project_name, df in dataset_df.items():
    # Prendre le DataFrame SANS la colonne 'Release version'
    df_counts = dataset_counts[project_name]
    corr_matrix = df_counts.corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
    plt.title(f'Correlation Heatmap of Smells: {project_name}')
    plt.show()
```

```python
# Create a global Heatmap by concatening all the DataFrames
combined_df = pd.concat(dataset_counts.values(), axis=0, ignore_index=True)
global_corr = combined_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(global_corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Global Correlation Heatmap of Smells (All Projects Combined)')
plt.show()
```
